#Exercise_1. Draw a line plot that shows the growth of air temperature over a 12-month period.The horizontal axis is the months of the year and the vertical axis is the temperature. Add an appropriate title. Specify the temperature value for each point (with 'marker='o ).

```python
import matplotlib.pyplot as plt

def plot_temperature_growth(months, temperatures):
    """
    Draws a line plot for temperature growth over months.
    Parameters
    ----------
    months : list of str
        Names of the months (x-axis labels).
    temperatures : list of int or float
        Temperature values corresponding to each month (y-axis values).
    """
    plt.figure(figsize=(10, 6))

    # Plot line with markers
    plt.plot(months, temperatures, marker='o', linestyle='-', color='royalblue', linewidth=2)

    # Add title and labels
    plt.title("üå°Ô∏è Air Temperature Growth Over 12 Months", fontsize=14, fontweight='bold')
    plt.xlabel("Months of the Year", fontsize=12)
    plt.ylabel("Temperature (¬∞C)", fontsize=12)

    # Show grid for better readability
    plt.grid(True, linestyle='--', alpha=0.6)

    # Annotate temperature values on each point
    for month, temp in zip(months, temperatures):
        plt.text(month, temp + 0.5, f"{temp}¬∞C", ha='center', fontsize=9, color='darkred')

    # Make layout tight
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    # Example dataset
    months_list = [
        "Jan", "Feb", "Mar", "Apr", "May", "Jun",
        "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"
    ]
    temperature_values = [4, 6, 10, 15, 20, 25, 30, 29, 24, 18, 10, 6]

    plot_temperature_growth(months_list, temperature_values)

```

##Exercise_2. Draw a bar chart that shows the number of students in three different majors over five consecutive years. Display two data series on one chart (for example, major A and major B). The colors of the bars should be different and plt.legend should be used.

```python
# üìä Bar chart showing the number of students in three majors over five consecutive years.

import matplotlib.pyplot as plt
import numpy as np

def plot_students_bar_chart(years, major_A, major_B, major_C):
    """
    Draws a bar chart for the number of students in three majors over given years.
    Parameters:
    -----------
    years : list of str
        Years for the x-axis (e.g., ["2019", "2020", ...]).
    major_A : list of int
        Number of students in Major A for each year.
    major_B : list of int
        Number of students in Major B for each year.
    major_C : list of int
        Number of students in Major C for each year.
    """

    # Calculate positions of the bars on the x-axis
    x = np.arange(len(years))
    width = 0.25  # width of each bar

    # Plot bars for each major
    plt.bar(x - width, major_A, width, label="Major A", color="skyblue")
    plt.bar(x, major_B, width, label="Major B", color="salmon")
    plt.bar(x + width, major_C, width, label="Major C", color="lightgreen")

    # Add labels and title
    plt.xlabel("Year")
    plt.ylabel("Number of Students")
    plt.title("Number of Students in Three Majors")
    plt.xticks(x, years)  # set x-axis tick labels
    plt.legend()          # add legend
    plt.grid(axis="y", linestyle="--", alpha=0.7)  # add horizontal grid lines

    # Adjust layout and display the plot
    plt.tight_layout()
    plt.show()


# ========================
# üéØ Example usage
# ========================
years = ["2019", "2020", "2021", "2022", "2023"]
major_A = [120, 135, 150, 160, 170]
major_B = [100, 115, 130, 145, 155]
major_C = [80, 95, 110, 125, 140]

plot_students_bar_chart(years, major_A, major_B, major_C)
```

#Exercise_3. Show the share of each of the 5 product types in a store's total sales in a month. Show the percentage of each section in the labels. Choose your own colors. Add a title "Sales share of each product".

```python
"""
This script draws a pie chart that shows the share of each of the
5 product types in a store's total monthly sales. The percentage of
each section is shown on the chart. The chart title is:
"Sales share of each product".
"""

import matplotlib.pyplot as plt

def plot_sales_share(product_names, sales_values, colors=None, save_path=None):
    """
    Plot a pie chart showing the sales share (percentage) of each product type.

    Parameters
    ----------
    product_names : list of str
        Names of the product types (e.g., ["Beverages", "Snacks", ...]).
    sales_values : list of float
        Sales values for each product type.
    colors : list of str, optional
        Custom colors for slices (hex codes or names). Default uses matplotlib palette.
    save_path : str, optional
        If provided, the plot will be saved as an image (e.g. "sales_share.png").
    """

    # Create pie chart
    plt.figure(figsize=(8, 8))
    plt.pie(
        sales_values,
        labels=product_names,
        autopct="%.1f%%",        # show percentage on slices
        colors=colors,
        startangle=90,
        counterclock=False,
        wedgeprops={"edgecolor": "white", "linewidth": 1},
        textprops={"fontsize": 10}
    )

    # Add title
    plt.title("Sales share of each product", fontsize=14, fontweight="bold")

    # Equal aspect ratio ensures pie is drawn as a circle
    plt.axis("equal")


    # Show chart
    plt.show()


if __name__ == "__main__":
    # Example usage with 5 product types
    products = ["Beverages", "Snacks", "Household", "Cosmetics", "Stationery"]
    sales = [12000, 8000, 5000, 3000, 2000]  # example monthly sales values
    my_colors = ["#ff9999", "#66b3ff", "#99ff99", "#ffcc99", "#c2c2f0"]

    plot_sales_share(products, sales, colors=my_colors, save_path="exercise_3_sales_share.png")
```
