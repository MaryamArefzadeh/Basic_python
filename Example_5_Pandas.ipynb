##Join a dataset with 90,000 unique rows with columns for first name, last name, email, contact number, ... randomly to the Hotel Booking dataset.

```python
!pip install faker
```

```python
import pandas as pd
from faker import Faker

fake = Faker()
Faker.seed(42)

n = 90000
personal_data = []

for _ in range(n):
    personal_data.append({
        "first_name": fake.first_name(),
        "last_name": fake.last_name(),
        "email": fake.unique.email(),
        "phone": fake.unique.phone_number(),
        "credit_card": fake.unique.credit_card_number(),
    })

personal_df = pd.DataFrame(personal_data)
personal_df.head()
```

```python
#Horizontally concat two dataframes
random_people = personal_df.sample(n=len(hotel_bookings), replace=True, random_state=42).reset_index(drop=True)
random_people.head()
final_df = pd.concat([hotel_bookings.reset_index(drop=True), random_people], axis=1)
final_df.head()
```

```python
#The name of the person who made the highest ADR payment and the amount paid.
max_adr_row = final_df.loc[final_df['adr'].idxmax()]
max_adr_name = max_adr_row['first_name']
max_ADR= max_adr_row['adr']

print(f"The person with the highest ADR is {max_adr_name} with an ADR of {max_ADR:.2f}")
```

```python
# calculate the average daily rate `adr` for a guest staying at each hotel
adr_by_hotel = final_df.groupby('hotel').mean(numeric_only=True)['adr']
adr_by_hotel
```

```python
#Average of the total ADR column with two decimal places
average_adr = final_df['adr'].mean()
print(f"The average ADR is {average_adr:.2f}")
```

```python
top5_lastnames = final_df['last_name'].value_counts().head(5).index

print("Top 5 most frequent last names:", list(top5_lastnames))

top5_people = final_df[final_df['last_name'].isin(top5_lastnames)]

print(top5_people[['first_name', 'last_name']])
```

```python
# People who booked hotels with the largest number of babies and children
final_df['total_kids'] = final_df['children'] + final_df['babies']

max_kids = final_df['total_kids'].max()

max_kids_rows = final_df[final_df['total_kids'] == max_kids]

result = max_kids_rows[['first_name', 'last_name', 'children', 'babies']]

print(result)
```

```python
max_children = final_df['children'].max()
max_babies = final_df['babies'].max()

max_children_rows = final_df[final_df['children'] == max_children]
max_babies_rows = final_df[final_df['babies'] == max_babies]

result_children = max_children_rows[['first_name', 'last_name', 'children']]
result_babies = max_babies_rows[['first_name', 'last_name', 'babies']]

print(result_children)
print(result_babies)
```

```python
# Percentage of return per hotel
```

```python
# city_hotel
returning_customer_city_hotel = sum(final_df['is_repeated_guest'] == 1)
total_customer_city_hotel = final_df.value_counts('hotel')['City Hotel']

percentage_returning_customer_city_hotel = (
    returning_customer_city_hotel * 100 / total_customer_city_hotel
)
percentage_returning_customer_city_hotel.round(2)
print(f"{percentage_returning_customer_city_hotel:.2f}%")
```

```python
# resort_hotel
returning_customer_resort_hotel = sum(final_df['is_repeated_guest'] == 1)
total_customer_resort_hotel = final_df.value_counts('hotel')['Resort Hotel']

percentage_returning_customer_resort_hotel = (
    returning_customer_resort_hotel * 100 / total_customer_resort_hotel
)
percentage_returning_customer_resort_hotel.round(2)
print(f"{percentage_returning_customer_resort_hotel:.2f}%")
```
